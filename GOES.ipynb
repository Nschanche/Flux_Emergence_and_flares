```python
def combine_lightcurves_GOES(start_year, start_month, start_day, end_year, end_month, end_day, fname):
    day = datetime.datetime(start_year, start_month, start_day)
    end_time = datetime.datetime(end_year, end_month, end_day)
    delta_day = datetime.timedelta(days=1)
    base_lightcurve = get_lightcurves_GOES(day, delta_day)
    base_lightcurve = base_lightcurve.data
    day += delta_day
    while day <= end_time:
        day1 = day+delta_day
        day_str_slash = day.strftime('%Y/%m/%d')
        day1_str_slash = day1.strftime('%Y/%m/%d')
        print day_str_slash
        if (day_str_slash != '2012/03/22' and day_str_slash !='2012/10/23' and day_str_slash !='2012/10/24' and 
            day_str_slash !='2012/10/25' and day_str_slash !='2012/11/01' and day_str_slash !='2012/11/02' and 
            day_str_slash !='2012/11/03' and day_str_slash !='2012/11/04' and day_str_slash !='2012/11/08' and
           day_str_slash !='2012/11/14' and day_str_slash !='2012/11/15' and day_str_slash !='2014/12/05' and 
           day_str_slash !='2015/08/17' and day_str_slash !='2015/08/18' and day_str_slash !='2015/08/19' and 
            day_str_slash !='2016/03/11'):
            tr = TimeRange(day_str_slash, day1_str_slash)
            lightcurve = GOESLightCurve.from_timerange(tr) 
            base_lightcurve = base_lightcurve.append(lightcurve.data, ignore_index = False)
        day += delta_day
    base_lightcurve.to_csv(fname, date_format = '%Y-%m-%dT%X')
```

```python
def get_lightcurves_GOES(day, delta_day):
    day1 = day+delta_day
    day_str_slash = day.strftime('%Y/%m/%d')
    day1_str_slash = day1.strftime('%Y/%m/%d')
    print day_str_slash
    tr = TimeRange(day_str_slash, day1_str_slash)
    lightcurve = GOESLightCurve.from_timerange(tr) 
    return lightcurve
```
